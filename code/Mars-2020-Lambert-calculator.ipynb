# 2. LAMBERT PROBLEM FOR MARS-2020

### Calculating first and last dots of trajectory

```python
r_earth_real = (x_m2020_real_0**2 + y_m2020_real_0**2)**0.5
r_mars_real = (x_m2020_real_end**2 + y_m2020_real_end**2)**0.5

omega_e = math.atan(y_m2020_real_0/x_m2020_real_0)
omega_m = math.atan(y_m2020_real_end/x_m2020_real_end)

x_e = r_earth_real*math.cos(omega_e)
y_e = r_earth_real*math.sin(omega_e)

x_m = r_mars_real*math.cos(omega_m)
y_m = r_mars_real*math.sin(omega_m)
```

### Draw planets, orbits and Sun

```python
plt.figure(figsize=(10, 10))

planet_1_x = list()
planet_1_y = list()
for i in range(0, 361):
    x_dot = r_earth_real * np.cos(i*math.pi/180)
    y_dot = r_earth_real * np.sin(i*math.pi/180)
    planet_1_x.append(x_dot)
    planet_1_y.append(y_dot)
    

planet_2_x = list()
planet_2_y = list()
for i in range(0, 361):
    x_dot = r_mars_real * np.cos(i*math.pi/180)
    y_dot = r_mars_real * np.sin(i*math.pi/180)
    planet_2_x.append(x_dot)
    planet_2_y.append(y_dot)
    
    
plt.plot(planet_2_x, planet_2_y, color = 'red')
plt.plot(planet_1_x, planet_1_y, color = 'blue')
plt.scatter(x_e, y_e, color = 'blue')
plt.scatter(x_m, y_m, color = 'red')
plt.scatter(0, 0, color = 'yellow')
```

### Set radius-vectors

```python
r_1 = np.array([x_e, y_e])
r_2 = np.array([x_m, y_m])
D = r_2 - r_1
```

### Define the angular values

```python
cos_fi = np.dot(r_1, r_2) / (np.linalg.norm(r_2) * np.linalg.norm(r_1))
sin_fi = np.linalg.norm(np.cross(r_1, r_2)) / (np.linalg.norm(r_2) * np.linalg.norm(r_1))
v_p = np.cross(r_1, r_2)
v_sign = math.copysign(1, v_p)
sin_fi = sin_fi * v_sign

if math.copysign(1, sin_fi) >= 0:
    fi = math.acos(cos_fi)
if math.copysign(1, sin_fi) < 0:
    fi = 2*math.pi - math.acos(cos_fi)
```

### Set the initial approximation to solve Lambert Problem

```python
mu = 1.3272e11

a_min = (np.linalg.norm(r_1) + np.linalg.norm(r_2) + np.linalg.norm(D)) / 4
delta_min = math.acos(1-(np.linalg.norm(r_1) + np.linalg.norm(r_2) - np.linalg.norm(D))/(2*a_min))

t_pz = a_min**(1.5)*(math.pi - math.copysign(1, sin_fi)*(delta_min - math.sin(delta_min)))/np.sqrt(mu)
```

### Calculating the time of mission

```python
t_p = 256 * 24*60*60

while t_pz < t_p:
    a_min = a_min + 100
    eps_v = math.acos(1-(np.linalg.norm(r_1) + np.linalg.norm(r_2) + np.linalg.norm(D))/(2*a_min))
    delta_v = math.acos(1-(np.linalg.norm(r_1) + np.linalg.norm(r_2) - np.linalg.norm(D))/(2*a_min))
    
    if t_p < t_pz:
        eps = eps_v
    if t_p > t_pz:
        eps = 2*math.pi - eps_v
    if t_p == t_pz:
        eps = math.pi
    
    if fi < math.pi:
        delta = delta_v
    if fi > math.pi:
        delta = - delta_v
    if fi == math.pi:
        delta = 0
    t_pz = (a_min**1.5)*(eps - delta - (math.sin(eps) - math.sin(delta)))/np.sqrt(mu)
```

### Define parameters of interplanetary flight

```python
p = np.linalg.norm(r_1) * np.linalg.norm(r_2) * ((math.sin(fi/2))**2) /(a_min * (math.sin((eps - delta)/2))**2)
ex = np.sqrt(1 - p / a_min)

cos_teta_1 = (p - np.linalg.norm(r_1)) / (ex * np.linalg.norm(r_1))
cos_teta_2 = (p - np.linalg.norm(r_2)) / (ex * np.linalg.norm(r_2))
sin_teta_1 = (math.cos(fi) * cos_teta_1 - cos_teta_2) / math.sin(fi)

if math.copysign(1, sin_teta_1) >= 0:
    teta_1 = math.acos(cos_teta_1)
if math.copysign(1, sin_teta_1) < 0:
    teta_1 = 2 * math.pi - math.acos(cos_teta_1)
    
teta_2 = teta_1 + fi

lambda0 = math.acos(x_e*(1+ex*math.cos(teta_1))/p) - teta_1
lambda02 = math.asin(y_e*(1+ex*math.cos(teta_1))/p) - teta_1
```

### Data of calculated trajectory to plot

```python
Lambert_transfer_x = list()
Lambert_transfer_y = list()

tetas = np.linspace(int(teta_1 * 180 / math.pi),int(teta_2 * 180 / math.pi),1000)

for i in range(0,1000):
    teta_i = tetas[i]*math.pi / 180
    rg_i = p / (1 + ex * math.cos(teta_i))
    Lambert_x = rg_i * math.cos(teta_i + lambda0)
    Lambert_y = rg_i * math.sin(teta_i + lambda0)
    Lambert_transfer_x.append(Lambert_x)
    Lambert_transfer_y.append(Lambert_y)
    
phi = -105*math.pi/180
Lambert_transfer_x_new = list()
Lambert_transfer_y_new = list()

for i in range(len(Lambert_transfer_x)):
    Lambert_transfer_x_new_i = Lambert_transfer_x[i]*math.cos(phi) - Lambert_transfer_y[i]*math.sin(phi)
    Lambert_transfer_y_new_i = Lambert_transfer_y[i]*math.cos(phi) + Lambert_transfer_x[i]*math.sin(phi)
    Lambert_transfer_x_new.append(Lambert_transfer_x_new_i)
    Lambert_transfer_y_new.append(Lambert_transfer_y_new_i)
```

### Plotting data

```python
plt.figure(figsize=(12, 12))

plt.plot(planet_2_x, planet_2_y, color = 'red')
plt.plot(planet_1_x, planet_1_y, color = 'blue')
plt.plot(Lambert_transfer_x_new, Lambert_transfer_y_new, label = 'Mars (calculated)', color = 'black')
plt.scatter(x_m, y_m, label = 'Mars (real data)', color = 'red')
plt.scatter(x_e, y_e, label = 'Earth', color = 'blue')
plt.scatter(0, 0,  label = 'Sun', color = 'yellow')
plt.plot(x_m2020_real, y_m2020_real, color = 'green')
plt.legend()
```

### Correct the angle of Mars-2020 relative to the sun

```python
r_lam = np.sqrt(np.array(Lambert_transfer_x)**2 + np.array(Lambert_transfer_y)**2)

angle_lam = list()

for i in range(len(Lambert_transfer_y)):
    if Lambert_transfer_x[i] > 0 and Lambert_transfer_y[i] >= 0:
        angle_lam.append(math.atan(Lambert_transfer_y[i]/Lambert_transfer_x[i]))
    if Lambert_transfer_x[i] < 0 and Lambert_transfer_y[i] >= 0:
        angle_lam.append(math.pi - math.atan(abs(Lambert_transfer_y[i]/Lambert_transfer_x[i])))
    if Lambert_transfer_x[i] < 0 and Lambert_transfer_y[i] < 0:
        angle_lam.append(math.pi + math.atan(abs(Lambert_transfer_y[i]/Lambert_transfer_x[i])))
    if Lambert_transfer_x[i] > 0 and Lambert_transfer_y[i] < 0:
        angle_lam.append(2*math.pi - math.atan(abs(Lambert_transfer_y[i]/Lambert_transfer_x[i])))
    if Lambert_transfer_x[i] == 0 and Lambert_transfer_y[i] > 0:
        angle_lam.append(math.pi/2)
    if Lambert_transfer_x[i] == 0 and Lambert_transfer_y[i] < 0:
        angle_lam.append(3*math.pi/2)

angle_lam = np.array(angle_lam)
```

###  Convertion of real and calculated radius data to a same size

#### Real:

```python
r_jpl = (x_m2020_real**2 + y_m2020_real**2)**0.5
angle_jpl = list()

for i in range(len(x_m2020_real.values)):
    if x_m2020_real.values[i] > 0 and y_m2020_real.values[i] >= 0:
        angle_jpl.append(math.atan(y_m2020_real.values[i]/data_m2020[2].values[i]))
    if x_m2020_real.values[i] < 0 and y_m2020_real.values[i] >= 0:
        angle_jpl.append(math.pi - math.atan(abs(y_m2020_real.values[i]/x_m2020_real.values[i])))
    if x_m2020_real.values[i] < 0 and y_m2020_real.values[i] < 0:
        angle_jpl.append(math.pi + math.atan(abs(y_m2020_real.values[i]/x_m2020_real.values[i])))
    if x_m2020_real.values[i] > 0 and y_m2020_real.values[i] < 0:
        angle_jpl.append(2*math.pi - math.atan(abs(y_m2020_real.values[i]/x_m2020_real.values[i])))
    if x_m2020_real.values[i] == 0 and y_m2020_real.values[i] > 0:
        angle_jpl.append(math.pi/2)
    if x_m2020_real.values[i] == 0 and y_m2020_real.values[i] < 0:
        angle_jpl.append(3*math.pi/2)
angle_jpl = np.array(angle_jpl)
```

#### Calculated:

```python
cos = np.vectorize(lambda t: math.cos(t))(angle_lam)
sin = np.vectorize(lambda t: math.sin(t))(angle_lam)
X = r_lam*cos
Y = r_lam*sin

A = np.array([
    [X[0]**2, X[0] * Y[0], Y[0]**2, X[0], Y[0]],
    [X[250]**2, X[250] * Y[250], Y[250]**2, X[250], Y[250]],
    [X[500]**2, X[500] * Y[500], Y[500]**2, X[500], Y[500]],
    [X[750]**2, X[750] * Y[750], Y[750]**2, X[750], Y[750]],
    [X[999]**2, X[999] * Y[999], Y[999]**2, X[999], Y[999]]
    
    ])
B = np.array([-1, -1, -1, -1, -1]).reshape((5,1))
x = solve(A,B)

d_fi = fi / 1000
fi_0 = math.atan(Lambert_transfer_y[0] / Lambert_transfer_x[0])

ans_list = list()
for j in range(0,1000): 
    b = x[3] * math.cos(angle_jpl[j]) + x[4] * math.sin(angle_jpl[j])
    a = x[0] * math.cos(angle_jpl[j]) * math.cos(angle_jpl[j]) + x[1]*math.cos(angle_jpl[j])*math.sin(angle_jpl[j]) + x[2]*math.sin(angle_jpl[j])*math.sin(angle_jpl[j])
    c = 1
    D = b**2 - 4 * a * c
    ans = (-b - np.sqrt(D)) / (2*a)
    ans = ans[0]
    ans_list.append(ans)
```
